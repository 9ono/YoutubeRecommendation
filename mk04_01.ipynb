```python
from collections import Counter, OrderedDict

def count_items(l):
    '''Return ordered dictionary of counts of objects in `l`'''
    
    # counter 객체를 생성한다.
    counts = Counter(l)
    
    # 가장 빈번하게 나온 video를 dictionary에서 정렬
    counts = sorted(counts.items(), key = lambda x: x[1], reverse = True)
    counts = OrderedDict(counts)
    
    return counts
```

```python
import random

random.seed(100)

def generate_batch(pairs, n_positive = 50, negative_ratio = 1.0, classification = False):
    batch_size = n_positive * (1 + negative_ratio)
    batch = np.zeros((batch_size, 3))
    
    #print(batch_size)
    
    '''분류 모델을 사용한다면 0으로 negative 값을 준다.
       이외에는 -1을 negative 값을 준다.'''
    if classification:
        neg_label = 0
    else:
        neg_label = -1
    
    '''generator 생성'''
    while True:
        '''batch size 중에서 n_positive만큼 랜덤으로 뽑아 positive값을 준다.'''
        for idx, (video_id, attribute_id) in enumerate(random.sample(pairs, n_positive)):
            batch[idx, :] = (video_id, attribute_id, 1)

        idx += 1
        
        '''나머지 batch size를 채우기 위해 랜덤으로 매치한 뒤에 negative 값을 준다.'''
        while idx < batch_size:
            
            # 무작위 선택
            random_video = random.randrange(len(videos))
            random_attribute = random.randrange(len(attributes))
            
            # positive 값을 가지는지 확인한다.
            if (random_video, random_attribute) not in pairs_set:
                
                # batch에 추가한다.
                batch[idx, :] = (random_video, random_attribute, neg_label)
                idx += 1
                
        # 무작위로 섞는다.
        np.random.shuffle(batch)
        yield {'video': batch[:, 0], 'attribute': batch[:, 1]}, batch[:, 2]
```

```python
from keras.layers import Input, Embedding, Dot, Reshape, Dense
from keras.models import Model

'''케라스 모듈 import'''
```

```python
def video_embedding_model(embedding_size = 100, classification = False):
    """Model to embed videos and tags using the functional API.
       Trained to discern if a link is present in a article"""
    
    # 1차원으로 두개의 입력을 받는다.
    video = Input(name = 'video', shape = [1])
    attribute = Input(name = 'attribute', shape = [1])
    
    # video를 embedding 한다.
    video_embedding = Embedding(name = 'video_embedding',
                               input_dim = len(video_index),
                               output_dim = embedding_size)(video)
    
    # attribute를 embedding 한다.
    attribute_embedding = Embedding(name = 'attribute_embedding',
                               input_dim = len(attribute_index),
                               output_dim = embedding_size)(attribute)
    
    # 내적하여 shape(1,1)으로 합친다.
    merged = Dot(name = 'dot_product', normalize = True, axes = 2)([video_embedding, attribute_embedding])
    
    # shape(1)로 변환한다.
    merged = Reshape(target_shape = [1])(merged)
    
    # classification에서만 실행
    if classification:
        merged = Dense(1, activation = 'sigmoid')(merged)
        model = Model(inputs = [video, attribute], outputs = merged)
        model.compile(optimizer = 'Adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    
    # Otherwise loss function is mean squared error
    else:
        model = Model(inputs = [video, attribute], outputs = merged)
        model.compile(optimizer = 'Adam', loss = 'mse')
    
    return model
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
plt.style.use('fivethirtyeight')
plt.rcParams['font.size'] = 15

def find_similar(name, weights, index_name = 'video', n = 20, least = False, return_dist = False, plot = False):
    '''비슷한 영상을 추천해 출력하는 함수'''
    
    if index_name == 'video':
        index = video_index
        rindex = index_video
    
    # video_index에 존재하지 않는 key를 거르기 위한 try문
    try:
        # 영상간의 내적 계산
        dists = np.dot(weights, weights[index[name]])
    except KeyError:
        print(f'{name} Not Found.')
        return
    
    # Sort 크기순으로 오름차순 정렬
    sorted_dists = np.argsort(dists)
    
    # 결과를 그래프로 표현 하위 n//2 개, 상위 n 개
    if plot:
        furthest = sorted_dists[:(n // 2)]
        closest = sorted_dists[-n-1: len(dists) - 1]
        items = [rindex[c] for c in furthest]
        items.extend(rindex[c] for c in closest)
        
        distances = [dists[c] for c in furthest]
        distances.extend(dists[c] for c in closest)
        
        colors = ['r' for _ in range(n //2)]
        colors.extend('g' for _ in range(n))
        
        data = pd.DataFrame({'distance': distances}, index = items)
        
        # 평행 그래프 그리기
        data['distance'].plot.barh(color = colors, figsize = (10, 8),
                                   edgecolor = 'k', linewidth = 2)
        plt.xlabel('Cosine Similarity');
        plt.axvline(x = 0, color = 'k');
        
        name_str = f'{index_name.capitalize()}s Most and Least Similar to'
        for word in name.split():            
            name_str += ' $\it{' + word + '}$'
        plt.title(name_str, x = 0.2, size = 28, y = 1.05)
        
        return None
    
    # 가장 비슷하지 않는 영상 추천
    if least:
        # dot 값이 가장 작은 것을 n까지 출력
        closest = sorted_dists[:n]
         
        print(f'{index_name.capitalize()}s furthest from {name}.\n')
        
    # 가장 비슷한 영상 추천
    else:
        # 뒤에서부터 dot 값이 높은 것들을 출력
        closest = sorted_dists[-n:]
        
        if return_dist:
            return dists, closest
        
        
        print(f'{index_name.capitalize()}s closest to {name}.\n')
        
    if return_dist:
        return dists, closest
    
    
    max_width = max([len(rindex[c]) for c in closest])
    
    for c in reversed(closest):
        print(f'{index_name.capitalize()}: {rindex[c]:{max_width + 2}} Similarity: {dists[c]:.{2}}')
```

```python
%matplotlib inline
plt.style.use('fivethirtyeight')
plt.rcParams['font.size'] = 15

def find_similar_list(name, weights, index_name = 'video', n = 20, least = False, return_dist = False, plot = False):
    '''비슷한 영상을 추천리스트로 만들어 return하는 함수'''
    
    if index_name == 'video':
        index = video_index
        rindex = index_video
    
    # video_index에 존재하지 않는 key를 거르기 위한 try문
    try:
        # 영상간의 내적 계산
        dists = np.dot(weights, weights[index[name]])
    except KeyError:
        print(f'{name} Not Found.')
        return
    
    # Sort 크기순으로 오름차순 정렬
    sorted_dists = np.argsort(dists)
    
    # 결과를 그래프로 표현 하위 n//2 개, 상위 n 개
    if plot:
        furthest = sorted_dists[:(n // 2)]
        closest = sorted_dists[-n-1: len(dists) - 1]
        items = [rindex[c] for c in furthest]
        items.extend(rindex[c] for c in closest)
        
        distances = [dists[c] for c in furthest]
        distances.extend(dists[c] for c in closest)
        
        colors = ['r' for _ in range(n //2)]
        colors.extend('g' for _ in range(n))
        
        data = pd.DataFrame({'distance': distances}, index = items)
        
        # 평행 그래프 그리기
        data['distance'].plot.barh(color = colors, figsize = (10, 8),
                                   edgecolor = 'k', linewidth = 2)
        plt.xlabel('Cosine Similarity');
        plt.axvline(x = 0, color = 'k');
        
        name_str = f'{index_name.capitalize()}s Most and Least Similar to'
        for word in name.split():            
            name_str += ' $\it{' + word + '}$'
        plt.title(name_str, x = 0.2, size = 28, y = 1.05)
        
        return None
    
    # 가장 비슷하지 않는 영상 추천
    if least:
        # dot 값이 가장 작은 것을 n까지 출력
        closest = sorted_dists[:n]
         
        #print(f'{index_name.capitalize()}s furthest from {name}.\n')
        
    # 가장 비슷한 영상 추천
    else:
        # 뒤에서부터 dot 값이 높은 것들을 출력
        closest = sorted_dists[-n:]
        
        if return_dist:
            return dists, closest
        
        
        #print(f'{index_name.capitalize()}s closest to {name}.\n')
        
    if return_dist:
        return dists, closest
    
    
    max_width = max([len(rindex[c]) for c in closest])
    
    rec_l = []
    # Print the most similar and distances
    for c in reversed(closest):
        rec_l.append(rindex[c])
    return rec_l
```

```python
def my_recall(answer_list = [], prediction_list = []):
    count = 0
    for i in prediction_list:
        if i in answer_list:
            count += 1
    length = 0
    
    for i in answer_list:
        if i != None:
            length += 1
    
    if length == 0:
        recall_result = 0.0 
    else:
        recall_result = count/length
        
        
    return recall_result
```

## embedding 결과 그래프로 확인하기

```python
from sklearn.manifold import TSNE
#from umap import UMAP
```

```python
def extract_weights(name, model):
    """Extract weights from a neural network model"""
    
    # Extract weights
    weight_layer = model.get_layer(name)
    weights = weight_layer.get_weights()[0]
    
    # Normalize
    weights = weights / np.linalg.norm(weights, axis = 1).reshape((-1, 1))
    return weights
```

```python
video_weights_class = extract_weights('video_embedding', model)
video_weights_class.shape
```

```python
attribute_weights_class = extract_weights('attribute_embedding', model)
```

```python
def reduce_dim(weights, components = 3, method = 'tsne'):
    """Reduce dimensions of embeddings"""
    if method == 'tsne':
        return TSNE(components, metric = 'cosine').fit_transform(weights)
    elif method == 'umap':
        # Might want to try different parameters for UMAP
        return UMAP(n_components=components, metric = 'cosine', 
                    init = 'random', n_neighbors = 5).fit_transform(weights)
```
